# Opinion score

Datasets for training:
- https://www.kaggle.com/clmentbisaillon/fake-and-real-news-dataset (true/fake.csv)
- https://www.kaggle.com/ruchi798/source-based-news-classification (news_articles.csv)

Datasets for testing:
- https://www.kaggle.com/snapcrack/all-the-news

```python
import pandas as pd
```

```python
# load raw datasets
true = pd.read_csv('./training/true.csv')
true
```

```python
fake = pd.read_csv('./training/fake.csv')
fake
```

```python
news = pd.read_csv('./training/news_articles.csv')
news
```

```python
from gensim.utils import simple_preprocess
```

```python
true.title = true.title.apply(lambda x: ' '.join(simple_preprocess(x)))
true['label'] = 'fact'
true = true[['title', 'label']]
true
```

```python
fake.title = fake.title.apply(lambda x: ' '.join(simple_preprocess(x)))
fake['label'] = 'opinion'
fake = fake[['title', 'label']]
fake
```

```python
news.dropna(subset=['label'], inplace=True)
news.title = true.title.apply(lambda x: ' '.join(simple_preprocess(x)))
news.label = news.label.apply(lambda x: 'fact' if x == 'Real' else 'opinion')
news = news[['title', 'label']]
news
```

```python
# combine all three
df = pd.concat([true, fake, news]).reset_index(drop=True)
df
```

```python
df.groupby('label').count()
```

```python
# output to fasttext compatible format
# __label__{label} {title}
# with 20% into validation dataset
import random

tpl = '__label__{label} {title}\n'

with open('train.txt', 'w') as tf, open('valid.txt', 'w') as vf:
    for row in df.itertuples():
        line = tpl.format(**row._asdict())
        if random.random() < 0.2:
            vf.write(line)
        else:
            tf.write(line)
```

```python
import fasttext

# For auto-tune (automatic hyperparameter optimization) by default, the search will take 5 minutes
model = fasttext.train_supervised(input='train.txt', autotuneValidationFile='valid.txt')
```

```python
model.predict('Trump ally charged with illegal foreign lobbying')
```

```python
model.predict(' '.join(simple_preprocess('Trump ally charged with illegal foreign lobbying')))
```

```python
# persist the model for reuse
model.save_model('./opinion.bin')
```

```python
import os
```

```python
# import testing datasets and cleanup
data = []
with os.scandir('./testing') as it:
    for entry in it:
        if entry.is_file():
            data.append(pd.read_csv(entry.path))

testing = pd.concat(data)
testing
```

```python
# take only what we need
testing = testing[['title', 'publication', 'date', 'content']].dropna(subset=['title'])
testing
```

```python
# apply our model
def add_score(row):
    title = ' '.join(simple_preprocess(row.title))  # normalize
    label, score = model.predict(title)  # predict
    return pd.Series({
        **row,
        'label': label[0].replace('__label__', ''),
        'score': score[0],
    })

testing = testing.apply(add_score, axis=1)
testing
```
