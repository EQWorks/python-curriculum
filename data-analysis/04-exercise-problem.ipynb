```python
import pandas as pd

# CSV export from https://console.locus.place/130/13799/builder?aws_id=8952cedaf7274ecfa23e210976714636
# VWI Report Exports - Download local report
df = pd.read_csv('../data/3479_vwi_local.csv')
df
```

```python
# populate `fsa` column from `address_postal` column
df['fsa'] = df['address_postal'].str.slice(0, 3)
df
```

```python
from datetime import datetime

format = '%Y-%m-%dT%H:%M:%S'

def get_days(x):
    start = datetime.strptime(x['start'], format)
    end = datetime.strptime(x['end'], format)
    return (end - start).days
    
df['days'] = df.apply(get_days, axis=1)
# filter out non-weekly rows
df = df.query('days <= 7')
df = df.reset_index()
df
```

```python
from datetime import datetime

# populate `week` column from `start` column
def gen_week(x):
    return datetime.strptime(x, '%Y-%m-%dT%H:%M:%S').isocalendar()[1]

df['week'] = df['start'].map(lambda x: datetime.strptime(x, '%Y-%m-%dT%H:%M:%S').isocalendar()[1])
df
```

```python
# group by fsa
stats = ['visits', 'unique_visitors', 'single_visit', 'multi_visit']
aggs = ['sum', 'mean', 'median', 'std']

gb = df.groupby(['fsa', 'week']).agg({k: aggs for k in stats})
gb
```

```python
# create pivot view with the index on `fsa` and the column on `week` for mean visits values
mean_vis_by = gb.reset_index().pivot(index='fsa', columns='week', values=('visits', 'mean'))
mean_vis_by
```
